# ANALISIS HARGA PRODUK PERTANIAN DI NEGARA ASIA TENGGARA

## RUMUSAN MASALAH

> Sektor pertanian masih menjadi ladang bisnis yang menguntungkan;

> Saat ini sudah banyak perusahaan Start-Up yang mulai menjajaki bisnis di sektor pertanian;

> Harga produk segar pertanian di negara-negara Asia Tenggara masih dinilai tidak menguntungkan bagi pelaku usaha terutama petani.


## TUJUAN

> Mengetahui perbandingan harga beberapa produk pertanian di negara asia tenggara dari tahun 2000 sampai 2018;

> Mengetahui pengaruh produksi terhadap harga beberapa bahan pertanian di negara asia tenggara dari tahun 2000 sampai 2018;

> Memprediksi harga produk pertanian sebagai bahan pertimbangan dalam memulai bisnis pertanian.

## DATASET

> Dataset yang digunakan adalah data set yang diambil dari situs resmi WHO "http://www.fao.org/faostat/"

> Dataset ini merupakan gabungan dari beberapa dataset yang ada dalam situs resmi WHO

> Dataset yang diagabungkan anatara lain : 
    Dataset produksi produk tanaman dan produk peternakan, 
    Dataset targa produk tanaman dan peternakan, 
    Dataset total import produk tanaman dan produk peternakan,
    Dataset total export produk tanaman dan produk peternakan,
    Dataset perubahan iklim setiap tahun di Negara Asia Tenggara
    

### Penjelasan Kolom Dataset

1. Domain
    Terdapat dua domain yaitu produk tanaman dan produk peternakan hewan 
2. Area
    Terdiri dari negara-negara yang ada di kawasan asia tenggara
3. Element
    Kolom ini menunjukkan asal dari angka yang didapatkan untuk kolom Production_Value
4. Item
    kolom jenis-jenis item apa saja yang akan dianalisa 
5. Year
    tahun didapatkannnya data
6. Prod_Unit
    satuan yang yang mengacu pada angka Production_Value
7. Production_Value
    Angka yang menunjukkan berapa jumlah hasil produk pertanian
8. Price_Unit
    satuan dari harga yang didapatkan pada item
9. Price_Value
    angka yang menunjukkan berapa harga jual dari item dan juga index harga dari sebuah item
10. Import_Unit
    satuan yang mengacu pada jumlah import suatu item
11. Import_Q
    angka yang menunjukkan jumlah import dari suatu item
12. Export_Unit
    satuan yang mengacu pada jumlah export suatu item
13. Export_Q
    angka yang menunjukkan jumlah import dari suatu item
14. Temp_Change_Unit
    satuan yang mengacu pada perubahan temperatur dari suatu negara
15. Temp_Change_Value
    angka yang menunjukkan berapa perubahan temperatur yang dialami sebuah negara
